```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
def plot(func, start=-10.0, end=10.0):
    x = np.linspace(start, end, 100)
    y = func(x)
    plt.plot(x, y)
```

# Sigmoid

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
plot(sigmoid)
```

# ReLU

```python
def relu(x):
    x = x*(x>0)
    x[x>=0] = x
    return x
```

```python
plot(relu)
```

# Softmax

```python
def softmax(x):
    e_x = np.exp(x-np.max(x))
    return e_x / e_x.sum(axis=0)

np.random.seed(101)
p = 2.5 * np.random.randn(2, 4) + 3
display(p)
plt.figure()
plt.imshow(p, cmap='gray')

s = softmax(p)
display(s)
plt.figure()
plt.imshow(s, cmap='gray')

plt.figure()
plot(softmax)
```
